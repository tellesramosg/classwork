```python
import pandas as pd

df = pd.read_csv("./data/french-ambs.csv")
df.head()
```

```python
for col in df.columns:
    print(f"{col} is numeric?: {pd.api.types.is_numeric_dtype(df[col])}")
```

```python
from random import choice, seed

seed(41)

# df["year"] = f"{df["year"]}-{choice(range(1,13)):02d}-{choice(range(1,29))}"
df["date_start"] = df["year"].apply(lambda x: f"{x}-{choice(range(1,13)):02d}-{choice(range(1,29)):02d}")

df.head()
```

```python
# drop the column date_start
df = df.drop(columns=["date_start"])
```

```python
import numpy as np
#make year_start better by using named fuction

def get_year_start(row):
    rows = df.loc[
        (df["name"] == row["name"]) &
        (df["place"] == row["place"]) &
        (df["year"] == row["year"]-1)
    ]
    if len(rows) > 0:
        return np.nan
    date_string = f"{row["year"]}-{choice(range(1,13)):02d}-{choice(range(1,29)):02d}"
    return date_string
   
df["date_start"] = df.apply(get_year_start, axis=1) # on a df, must pass axis
df.head()
```

```python
def get_year_end(row):
    rows = df.loc[
        (df["name"] == row["name"]) &
        (df["place"] == row["place"]) &
        (df["year"] == row["year"]+1)
    ]
    if len(rows) > 0:
        return np.nan
    date_string = f"{row["year"]}-{choice(range(1,13)):02d}-{choice(range(1,29)):02d}"
    return date_string
   
df["date_end"] = df.apply(get_year_end, axis=1) # on a df, must pass axis
df.head()
```

```python
# bfill, ffill

df = df.sort_values(["place","name", "year"], ascending=True).reset_index()
df.head()
```

```python
df["date_start"] = df["date_start"].ffill()
df["date_end"] = df["date_end"].bfill()
df.head()
```

```python
import datetime
fmt = "%Y-%M-%d"

df["date_start"] = df["date_start"].apply(lambda x: datetime.datetime.strptime(x, fmt))
df["date_end"] = df["date_end"].apply(lambda x: datetime.datetime.strptime(x, fmt))
df.head()
```

```python
df.groupby(["place"]).count()
```

```python
df.groupby(["place"]).agg(
    {
        "place": "count",
        "year": "min",
    }
)
```

```python
df.groupby(["place"]).agg(
    place_years=("place", "count"),
    max=("year", "max"),
    min=("year", "min"),
    num_ambs=("name", "nunique"),
    ambs=("name", "unique")
).reset_index()
```
