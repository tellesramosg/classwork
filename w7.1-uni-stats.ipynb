```python
import pandas as pd

url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/soccer-players.csv"
df = pd.read_csv(url)

df.head()
```

```python
def convert_height(cm) -> float:
    """
    convert cm to feet
    
    Args:
        cm (int): the Num of cm
    Returns:
        feet (floa): num of feet
    """
    return (cm * 0.393701) / 12
```

```python
df["height_in_ft"] = df["height_in_cm"].apply(convert_height)
df["height_in_ft"]
```

```python
def eda(df) -> None:
    """
    Evaluate basic statistics on a Dataframe
    
    Args:
        df (Pandas.DataFrame): DataFrame to ecaluate
    Returns:
        None
    """
    print(f"{'='*5}DF Shape: {df.shape}{'='*5}")
    num_cols = []
    cat_cols = []
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            num_cols.append(col)
        else:
            cat_cols.append(col)
            
    print(f"{'*'*3}Numeric Columns{'*'*3}")
    for col in num_cols:
        print(col)
        print(f"\tMin: {df[col.min()]}")
        print(f"\tMax: {df[col.max()]}")
```

```python
eda(df)
```
