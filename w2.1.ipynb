## Complex Varibale

```python
# string
string = "any sequence of characters" # strings use quotes or ' '


# never name variables this crappy way!
s = str("something else")

#snake casing, not hyohens, not NumberString, numberString
number_string = "67"
print (type(string))
print (type (s), s)

#Jupyter will print (display) the last time.
number_string
```

```python
# int
x = 1
y = int('3')

print (x)
print (type (y))

# aligning info to output
print("x plus y equals", x+y, ", but not f string")
print(f"x+y={x+y}, but with an f string")

#concatination is different from addition
string + str(x)
```

```python
# Float

#we can overwrite variable names
x = 1.0
y = float(42)

print( f"x = {x}, y = {y}, and y is a {type (y)}")
#notice what happens to 5
print(f"x + y = {x+y}, but y + 5 = {y+5}")
```

```python
# Boolean = true/false

is_true = True # False
is_false = bool(0)
print (is_true)
print(is_false)
```

```python
# Variables store the resolution of code.
outcome = x + y
print(outcome)
```

## Methods

```python
# string methods

print(type(string))
print(string.capitalize())
print(string.upper())
print(string.islower())
print(string.endswith("any"))
```

```python
# floats and ints have few methods
x = 10
y = 5
# +, -, *, /, % (modulus), ** (exponent)

print (x*y)
print (x/y)
print(x%y, "modulus")
print(x/3, "normal division")
print (x//3, "floor division")
```
