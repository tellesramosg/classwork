```python
for col in df.columns:
    print(col, pd.api.types.is_numeric_dtype(df[col]))
```

```python
# loop over rows
# iterrows returns a list

for idx, item in df.iterrows():
    print(item["place"])
```

```python
# itertuples() is faster
for item in df.itertuples():
    if item.place == "Denmark":
        df.at[item.Index, "place"] = "Den."
df["place"].unique() # get inuque values existing in column
```

```python
# England => Eng.
# If only changing one name, use .lic[]
df.loc[df["place"] == "England", "place"] = "Eng."
df["place"].unique()
```

```python
# .replace for miltiple

#hash = (
#    "holy Roman Empire": "HRE"
#   "Geneva": "Gen.'Gen."
#)
hash = {name:f"{name[:3]}." for name in df["place"].unique() if len(name) != 4}
df["place"] = df["place"].replace(hash)
df["place"].unique()
    
```

```python
# Pandas operators &(and), | (or), ~ (not)
df.loc[(df["place"] == "Den.") & (df["year"] < 1561) & (df["year"] > 1550)].count()
```

```python
# use ~
df.loc[~(df["place"] == "Den.") & (df["year"] < 1561) & (df["year"] > 1550)].count()
```

```python
imp_ambs = ["Eng.", "Ott.", "Ven."]

df.loc[(df["place"].isin(imp_ambs)) & (df["year"] < 1561) & (df["year"] > 1550)].count()
```

```python
for item in df["place"].unique():
    if item in imp_ambs:
        print(f"is {item} in ambs list? {item in imp_ambs}")
```
